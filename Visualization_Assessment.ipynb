#Line Graph

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

#Bar Graph

```python
import matplotlib.pyplot as plt

x = ['A','B','C','D']
y = [20,35,30,25]

plt.bar(x,y,color='green', edgecolor='black')
plt.title("Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#Histogram

```python
import matplotlib.pyplot as plt

x = [10,20,20,30,30,30,40,40,50]

plt.hist(x,bins=5,color='blue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

#Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [5,10,7,16,25]

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

#Pie Chart

```python
import matplotlib.pyplot as plt

x = ['Python','Java','C']
y = [40,35,25]

plt.pie(y,labels=x,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

#Box Plot

```python
import matplotlib.pyplot as plt

x = [5,10,15,20,25]
y = [30,35,40,45,50]

data = [x, y]

plt.boxplot(data)

plt.title("Box Plot")
plt.xlabel("Datasets")
plt.ylabel("Values")

plt.show()
```

#Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [[1,2,3],[4,5,6],[7,8,9]]
sns.heatmap(x,annot=True)
plt.title("Heatmap")
plt.show()
```
