#Seaborn

1.Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Student' : ['A','B','C','D','E'],
    'Marks' : [69,74,88,89,90]
}

df = pd.DataFrame(data)

sns.lineplot(x='Student', y='Marks', data=df)

plt.title("Student Marks Line Plot")
plt.xlabel("Student")
plt.ylabel("Marks")

plt.show()
```

#Kernal Density Estimate

```python


from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({
    0:'Setosa',
    1:'Versicolor',
    2:'Virginica'
})
sns.kdeplot(
    data=df[df['species'] == 'Versicolor'],
    x='sepal width (cm)',
    label='Versicolor',
    shade=True
)

plt.legend()

plt.title("KDE Plot of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Density")

plt.show()
```

#Customise the Seaborn

1.Adding titles and axis labels

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')

sns.scatterplot(
    x='sepal_length',
    y='petal_width',
    data=iris,
    hue='species'
)

plt.title('Scatter Plot of Sepal Length vs Petal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

2.Change the styles and grids using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
iris = sns.load_dataset('iris')
sns.boxplot(
    x='species',
    y='sepal_width',
    data=iris
)

plt.title('Box Plot of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')

plt.show()
```
